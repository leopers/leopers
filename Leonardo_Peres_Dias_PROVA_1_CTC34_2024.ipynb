## Questão 2 (valor 1 ponto)

## **Refutar:** "Todo subconjunto de uma linguagem regular é também uma linguagem regular".

Dica: partir de uma linguagem que é sabidamente não regular como $a^kb^k$, ou a linguagem dos palíndromos.

---

## Resposta questão 2

Tome Σ = {0,1}. Note que $L = 0^k1^k \subset Σ^*$ e que $Σ^*$ é trivialmente regular, pois é obtido como ($\bigcup_{p \in \Sigma} \{p\})^*$.

Lema: $L = 0^k1^k$ é não regular.

Prova:

Por contradição, suponha que $L$ é regular.

Seja $n$ como no Lema de Bombeamento.


Seja $x = 0^n 1^n$. \\
Então $x \in L$  \\
e $|x| = 2n \geq n$.

Pelo Lema de Bombeamento, existem cadeias $u, v, w$ tais que

  (i) $x = uvw$, \\
  (ii) $v \neq \varepsilon$, \\
  (iii) $|uv| \leq n$, \\
  (iv) $uv^k w \in L$ para todo $k \in \mathbb{N}$.


Seja $y$ o prefixo de $x$ com comprimento $n$. Isto é, $y$ são os primeiros $n$ símbolos de $x$ \\
Pela nossa escolha de $x$, $y = 0^n$. \\
Por (i) e (iii), $uv = 0^j$ para algum $j \in \mathbb{N}$ com $0 \leq j \leq n$. \\
Combinando com (ii), $v = 0^j$ para algum $j \in \mathbb{N}$ com $0 < j \leq n$.

Por (iv), $uv^2 w \in L$.  (#) \\
{Observação: Estamos escolhendo $k = 2$. De fato, qualquer $k \neq 1$ serve aqui.}

No entanto,
$
uv^2w = uvvw = 0^{n+j} 1^n
$
$\notin L$,  [definição de $L$; como $j > 0$, $n + j \neq n$] \\
o que contradiz (#).

Portanto, $L$ não é regular. $\square$

Como L não é regular, a afirmação foi refutada.
